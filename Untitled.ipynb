```python
%%writefile download.py
import argparse
import os
import urllib.request
import json
from tqdm import tqdm

BASE_URL = "http://kaldir.vc.in.tum.de/faceforensics/v3/"

DATASETS = {
    "original": "original_sequences/youtube",
    "Deepfakes": "manipulated_sequences/Deepfakes",
    "FaceSwap": "manipulated_sequences/FaceSwap",
    "Face2Face": "manipulated_sequences/Face2Face"
}

def download_file(url, path):
    os.makedirs(os.path.dirname(path), exist_ok=True)
    if not os.path.exists(path):
        try:
            urllib.request.urlretrieve(url, path)
        except:
            pass

def main():
    parser = argparse.ArgumentParser()
    parser.add_argument("output_path")
    parser.add_argument("--dataset", required=True)
    parser.add_argument("--compression", default="c23")
    parser.add_argument("--num_videos", type=int, default=50)
    args = parser.parse_args()

    dataset_path = DATASETS[args.dataset]

    print("Downloading:", args.dataset)

    filelist_url = BASE_URL + "misc/filelist.json"
    file_pairs = json.loads(urllib.request.urlopen(filelist_url).read().decode())

    filelist = []
    for pair in file_pairs:
        if args.dataset == "original":
            filelist += pair
        else:
            filelist.append(pair[0] + "_" + pair[1])
            filelist.append(pair[1] + "_" + pair[0])

    filelist = filelist[:args.num_videos]

    for fname in tqdm(filelist):
        video_url = "{}/{}/{}/videos/{}.mp4".format(
            BASE_URL.rstrip("/"),
            dataset_path,
            args.compression,
            fname
        )

        save_path = os.path.join(
            args.output_path,
            dataset_path,
            args.compression,
            "videos",
            fname + ".mp4"
        )

        download_file(video_url, save_path)

    print("Done:", args.dataset)

if __name__ == "__main__":
    main()
```

```python
import cv2
import os

def extract_frames(video_path, save_dir, max_frames=10):
    cap = cv2.VideoCapture(video_path)
    count = 0
    
    while True:
        ret, frame = cap.read()
        if not ret or count >= max_frames:
            break
        
        filename = os.path.basename(video_path).replace(".mp4", f"_{count}.jpg")
        cv2.imwrite(os.path.join(save_dir, filename), frame)
        count += 1
    
    cap.release()
```

```python
import cv2, os, random

def extract_frames(video_path, save_dir, max_frames=8):   # 🔥 reduce to 8
    
    cap = cv2.VideoCapture(video_path)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    
    if total_frames <= 0:
        return
    
    frame_ids = random.sample(range(total_frames), min(max_frames, total_frames))
    
    count = 0
    current = 0
    
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        
        if current in frame_ids:
            filename = os.path.basename(video_path).replace(".mp4", f"_{count}.jpg")
            cv2.imwrite(os.path.join(save_dir, filename), frame)
            count += 1
        
        current += 1
    
    cap.release()
```

```python
import numpy as np
import cv2
import os

def generate_fft(img_path, save_path):
    img = cv2.imread(img_path, 0)  # grayscale
    
    if img is None:
        return
    
    # FFT
    f = np.fft.fft2(img)
    fshift = np.fft.fftshift(f)
    
    magnitude = 20 * np.log(np.abs(fshift) + 1)
    
    # Normalize
    magnitude = cv2.normalize(magnitude, None, 0, 255, cv2.NORM_MINMAX)
    magnitude = magnitude.astype(np.uint8)
    
    cv2.imwrite(save_path, magnitude)
```

```python
from tqdm import tqdm

def process_fft(src, dst):
    os.makedirs(dst, exist_ok=True)
    
    files = os.listdir(src)
    
    for f in tqdm(files):
        generate_fft(
            os.path.join(src, f),
            os.path.join(dst, f)
        )
```

```python
import torch
import torch.nn as nn
import timm

class DualCNN(nn.Module):
    def __init__(self):
        super().__init__()
        
        # Spatial branch (faces)
        self.spatial = timm.create_model('efficientnet_b0', pretrained=True, num_classes=0)
        
        # Frequency branch (FFT)
        self.frequency = timm.create_model('efficientnet_b0', pretrained=True, num_classes=0)
        
        # Attention fusion
        self.attention = nn.Sequential(
            nn.Linear(1280 * 2, 512),
            nn.ReLU(),
            nn.Linear(512, 2),
            nn.Softmax(dim=1)
        )
        
        # Final classifier
        self.classifier = nn.Linear(1280 * 2, 1)

    def forward(self, x_spatial, x_freq):
        f1 = self.spatial(x_spatial)
        f2 = self.frequency(x_freq)
        
        combined = torch.cat([f1, f2], dim=1)
        
        attn = self.attention(combined)
        
        fused = combined * attn.repeat(1, combined.shape[1] // 2)
        
        out = self.classifier(fused)
        
        return out
```

```python
import timm
class DualCNN(nn.Module):
    def __init__(self):
        super().__init__()
        
        self.spatial = timm.create_model('mobilenetv3_small_100', pretrained=True, num_classes=0)
        self.frequency = timm.create_model('mobilenetv3_small_100', pretrained=True, num_classes=0)
        
        self.attention = nn.Sequential(
            nn.Linear(2048, 512),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(512, 2),
            nn.Softmax(dim=1)
        )
        
        # ⭐ UPDATED CLASSIFIER
        self.classifier = nn.Sequential(
            nn.Linear(2048, 256),
            nn.ReLU(),
            nn.Dropout(0.4),
            nn.Linear(256, 1)
        )

    def forward(self, x_spatial, x_freq):
        f1 = self.spatial(x_spatial)
        
        # ⭐ STRONG FFT BOOST
        f2 = self.frequency(x_freq) * 2.0
        
        combined = torch.cat([f1, f2], dim=1)
        
        attn = self.attention(combined)
        
        w1 = attn[:, 0].unsqueeze(1)
        w2 = attn[:, 1].unsqueeze(1)
        
        fused = torch.cat([f1 * w1, f2 * w2], dim=1)
        
        out = self.classifier(fused)
        
        return out
         
  
```

```python

def train_epoch(model, loader, optimizer, criterion, device):
    model.train()
    total_loss = 0
    
    for i, (face, fft, label) in enumerate(loader):
        if i % 50 == 0:
            print("Batch:", i)
        
        face, fft = face.to(device), fft.to(device)
        label = label.float().unsqueeze(1).to(device)  # 🔥 FIX
        
        optimizer.zero_grad()
        
        output = model(face, fft)  # ❌ no squeeze
        loss = criterion(output, label)
        
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item()
    
    return total_loss / len(loader)
```

```python
def evaluate(model, loader, criterion, device):
    model.eval()
    total_loss = 0
    correct = 0
    total = 0
    
    with torch.no_grad():
        for face, fft, label in loader:
            face, fft = face.to(device), fft.to(device)
            label = label.float().unsqueeze(1).to(device)  # 🔥 FIX
            
            output = model(face, fft)
            loss = criterion(output, label)
            
            total_loss += loss.item()
            
            preds = (torch.sigmoid(output) > 0.5).float()  # 🔥 FIX
            correct += (preds == label).sum().item()
            total += label.size(0)
    
    acc = correct / total
    return total_loss / len(loader), acc
```

```python
%%writefile download.py
import argparse
import os
import urllib.request
import json
from tqdm import tqdm
import random   # ✅ ADDED

BASE_URL = "http://kaldir.vc.in.tum.de/faceforensics/v3/"

DATASETS = {
    "original": "original_sequences/youtube",
    "Deepfakes": "manipulated_sequences/Deepfakes",
    "FaceSwap": "manipulated_sequences/FaceSwap",
    "Face2Face": "manipulated_sequences/Face2Face"
}

def download_file(url, path):
    os.makedirs(os.path.dirname(path), exist_ok=True)
    if not os.path.exists(path):
        urllib.request.urlretrieve(url, path)

def main():
    parser = argparse.ArgumentParser()
    parser.add_argument("output_path")
    parser.add_argument("--dataset", required=True)
    parser.add_argument("--compression", default="c23")
    parser.add_argument("--num_videos", type=int, default=50)
    args = parser.parse_args()

    dataset_path = DATASETS[args.dataset]

    print("Downloading:", args.dataset)

    filelist_url = BASE_URL + "misc/filelist.json"
    file_pairs = json.loads(urllib.request.urlopen(filelist_url).read().decode())

    filelist = []
    for pair in file_pairs:
        if args.dataset == "original":
            filelist += pair
        else:
            filelist.append("_".join(pair))

    # ✅ FIX (RANDOM SELECTION)
    random.shuffle(filelist)
    filelist = filelist[:args.num_videos]

    for fname in tqdm(filelist):
        video_url = "{}/{}/{}/videos/{}.mp4".format(
            BASE_URL.rstrip("/"),
            dataset_path,
            args.compression,
            fname
        )

        save_path = os.path.join(
            args.output_path,
            dataset_path,
            args.compression,
            "videos",
            fname + ".mp4"
        )

        download_file(video_url, save_path)

    print("Done:", args.dataset)

if __name__ == "__main__":
    main()
```

```python
import cv2, os

def extract_faces_from_frames(frame_dir, face_dir):
    os.makedirs(face_dir, exist_ok=True)
    face_detector = cv2.CascadeClassifier(
        cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
    )

    files = os.listdir(frame_dir)
    total = len(files)
    print(f"Processing {total} frames from {frame_dir} -> {face_dir}")

    for i, file in enumerate(files, 1):
        img_path = os.path.join(frame_dir, file)
        img = cv2.imread(img_path)
        if img is None:
            continue

        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = face_detector.detectMultiScale(gray, 1.3, 5)
        if len(faces) == 0:
            continue

        x, y, w, h = faces[0]
        face = img[y:y+h, x:x+w]
        save_path = os.path.join(face_dir, file)
        cv2.imwrite(save_path, face)

        if i % 100 == 0:
            print(f"{frame_dir}: {i}/{total} done")

extract_faces_from_frames("frames/train/real", "faces/train/real")
extract_faces_from_frames("frames/train/fake", "faces/train/fake")
extract_faces_from_frames("frames/val/real",   "faces/val/real")
extract_faces_from_frames("frames/val/fake",   "faces/val/fake")
```

```python
import cv2, os
import numpy as np

def compute_fft(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    f = np.fft.fft2(gray)
    fshift = np.fft.fftshift(f)
    magnitude = 20*np.log(np.abs(fshift)+1)
    # normalize to 0–255 for saving
    magnitude = cv2.normalize(magnitude, None, 0, 255, cv2.NORM_MINMAX)
    magnitude = magnitude.astype(np.uint8)
    return magnitude

def process_fft(input_dir, output_dir):
    os.makedirs(output_dir, exist_ok=True)

    for file in os.listdir(input_dir):
        img_path = os.path.join(input_dir, file)
        img = cv2.imread(img_path)

        if img is None:
            continue

        fft_img = compute_fft(img)
        fft_img = cv2.resize(fft_img, (224, 224))

        save_path = os.path.join(output_dir, file)
        cv2.imwrite(save_path, fft_img)
```

```python
import torch
from torch.utils.data import Dataset
from torchvision import transforms

transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize((224,224)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.ToTensor()
])

class DeepfakeDataset(Dataset):
    def __init__(self, face_dir, fft_dir, transform):
        self.face_paths = [os.path.join(face_dir, f) for f in os.listdir(face_dir)]
        self.face_paths = [p for p in self.face_paths if p.lower().endswith((".jpg", ".png", ".jpeg"))]
        self.fft_dir = fft_dir
        self.transform = transform

    def __len__(self):
        return len(self.face_paths)

    def __getitem__(self, idx):
        face_path = self.face_paths[idx]
        file = os.path.basename(face_path)

        fft_path = os.path.join(self.fft_dir, file)

        face = cv2.imread(face_path)
        fft = cv2.imread(fft_path)

        # fallback if fft missing
        if fft is None:
            fft = np.zeros_like(face)

        label = 1 if "fake" in face_path else 0

        face = self.transform(face)
        fft = self.transform(fft)

        return face, fft, torch.tensor(label)
```

```python
from torch.utils.data import DataLoader

train_real = DeepfakeDataset("faces/train/real", "fft/train/real", transform)
train_fake = DeepfakeDataset("faces/train/fake", "fft/train/fake", transform)

val_real = DeepfakeDataset("faces/val/real", "fft/val/real", transform)
val_fake = DeepfakeDataset("faces/val/fake", "fft/val/fake", transform)

train_dataset = train_real + train_fake
val_dataset = val_real + val_fake

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
val_loader   = DataLoader(val_dataset, batch_size=8, shuffle=False)

print("Train size:", len(train_dataset), "Val size:", len(val_dataset))
```

```python
import torch.nn as nn
import torchvision.models as models

class DualCNN(nn.Module):
    def __init__(self):
        super().__init__()

        self.spatial = models.mobilenet_v3_small(weights=models.MobileNet_V3_Small_Weights.IMAGENET1K_V1)
        self.frequency = models.mobilenet_v3_small(weights=models.MobileNet_V3_Small_Weights.IMAGENET1K_V1)

        self.spatial.classifier = nn.Identity()
        self.frequency.classifier = nn.Identity()

        self.classifier = nn.Sequential(
            nn.Linear(1152, 256),   # <- put the printed size here
            nn.ReLU(),
            nn.Linear(256, 1)
          )

    def forward(self, face, fft):
        f1 = self.spatial(face)
        f2 = self.frequency(fft)
        combined = torch.cat([f1, f2], dim=1)
        return self.classifier(combined)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = DualCNN().to(device)
```

```python
criterion = nn.BCEWithLogitsLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4, weight_decay=1e-5)

def train_epoch(model, loader, optimizer, criterion, device):
    model.train()
    total_loss = 0
    correct = 0
    total = 0

    for face, fft, label in loader:
        face, fft = face.to(device), fft.to(device)
        label = label.float().to(device)

        optimizer.zero_grad()
        output = model(face, fft)
        loss = criterion(output, label.unsqueeze(1))
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

        preds = (torch.sigmoid(output) > 0.5).squeeze()
        correct += (preds == label).sum().item()
        total += label.size(0)

    acc = correct / total
    return total_loss / len(loader), acc

def evaluate(model, loader, criterion, device):
    model.eval()
    total_loss = 0
    correct = 0
    total = 0

    with torch.no_grad():
        for face, fft, label in loader:
            face, fft = face.to(device), fft.to(device)
            label = label.float().to(device)

            output = model(face, fft)
            loss = criterion(output, label.unsqueeze(1))

            preds = (torch.sigmoid(output) > 0.5).squeeze()
            correct += (preds == label).sum().item()
            total += label.size(0)

            total_loss += loss.item()

    acc = correct / total
    return total_loss / len(loader), acc
```

```python
epochs = 25
best_val_acc = 0.0
patience = 7        # allows 3 bad epochs in a row
no_improve = 0

for epoch in range(epochs):
    print(f"\nEpoch {epoch+1}")

    train_loss, train_acc = train_epoch(model, train_loader, optimizer, criterion, device)
    val_loss, val_acc = evaluate(model, val_loader, criterion, device)

    print(f"Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f}")
    print(f"Val   Loss: {val_loss:.4f}, Val   Acc: {val_acc:.4f}")

    if val_acc > best_val_acc:
        best_val_acc = val_acc
        no_improve = 0              # reset because it improved
        torch.save(model.state_dict(), "best_model.pth")
        print("🔥 Best Model Saved! (val_acc = {:.4f})".format(best_val_acc))
    else:
        no_improve += 1
        print(f"No improvement for {no_improve} epoch(s)")

    if no_improve >= patience:
        print("Early stopping triggered!")
        break
```
